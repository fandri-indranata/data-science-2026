```python
import pandas as pd
import numpy as np

# Dataset Titanic
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Menampilkan 5 baris pertama
df.head()
```

```python
# Q1: Berapa jumlah baris dan kolom pada dataset Titanic? Kolom apa saja yang tersedia?
print("Jumlah Baris dan Kolom (Rows,Columns):",df.shape)
print("\nDaftar Kolom yang tersedia:")
print(df.columns.tolist())
```

Pada dataset titanic ini memiliki 891 baris dengan 12 kolom, kolom-kolom ini terdiri dari 'PassengerId', 'Survived', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin' dan 'Embarked'.

```python
# Q2: Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%) per kolom?
missing_count = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100
# Membuat DataFrame baru untuk menggabungkan jumlah dan persentase
missing_df = pd.DataFrame({
    'Jumlah_Missing': missing_count,
    'Persentase(%)': missing_percent.round(2)
})
# Hanya tampilkan kolom yang memiliki missing values
print("Kolom dengan Missing Values:")
print(missing_df[missing_df['Jumlah_Missing'] > 0])
```
