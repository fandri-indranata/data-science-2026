```python
import pandas as pd
import numpy as np

# Dataset Titanic
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Menampilkan 5 baris pertama
df.head()
```

```python
# Q1: Berapa jumlah baris dan kolom pada dataset Titanic? Kolom apa saja yang tersedia?
print("Jumlah Baris dan Kolom (Rows,Columns):",df.shape)
print("\nDaftar Kolom yang tersedia:")
print(df.columns.tolist())
```

Pada dataset titanic ini memiliki 891 baris dengan 12 kolom, kolom-kolom ini terdiri dari 'PassengerId', 'Survived', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin' dan 'Embarked'.

```python
# Q2: Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%) per kolom?
missing_count = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100
# Membuat DataFrame baru untuk menggabungkan jumlah dan persentase
missing_df = pd.DataFrame({
    'Jumlah_Missing': missing_count,
    'Persentase(%)': missing_percent.round(2)
})
# Hanya tampilkan kolom yang memiliki missing values
print("Kolom dengan Missing Values:")
print(missing_df[missing_df['Jumlah_Missing'] > 0])
```

Terdapat 3 kolom yang memiliki missing values: Age (177 data / 19,87%), Cabin (687 data / 77,10%), dan Embarked (2 data / 0,22%)."

```python
# Q3: Berapa persen penumpang yang selamat (Survived=1)? Gunakan operasi NumPy atau Pandas
print(f"Persentase penumpang selamat: {np.mean(df['Survived']) * 100:.2f}%")
```

Dengan menggunakan Numpy didapat hanya sekitar 38,38% penumpang yang selamat dari bencana tenggelamnya kapal Titanic.

```python
# Q4: Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa rata-rata usianya?
wanita_kelas1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]

# Menghitung jumlah dan rata-rata usia
jumlah_wanita_kelas1 = len(wanita_kelas1)
rata_usia = wanita_kelas1['Age'].mean()

print(f"Jumlah penumpang wanita kelas 1: {jumlah_wanita_kelas1} orang")
print(f"Rata-rata usia mereka: {rata_usia:.2f} tahun")
```

Pada data titanic terdapat 94 penumpang wanita di kelas 1. Rata-rata usia mereka adalah sekitar 34,61 tahun

```python
# Q5: Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas penumpang. Kelas mana yang paling tinggi?
tingkat_selamat_kelas = df.groupby('Pclass')['Survived'].mean() * 100

print("Tingkat Keselamatan per Kelas Penumpang (%):")
print(tingkat_selamat_kelas.round(2))

# Mengetahui kelas mana yang paling tinggi
kelas_tertinggi = tingkat_selamat_kelas.idxmax()
print(f"\nKelas dengan tingkat keselamatan tertinggi: Kelas {kelas_tertinggi}")
```

Kelas 1 memiliki tingkat keselamatan tertinggi (sekitar 62,96%), diikuti Kelas 2 (47,28%), dan Kelas 3 (24,24%). Ini menunjukkan bahwa penumpang di kelas utama (kaya) memiliki peluang selamat yang jauh lebih besar.

```python
# Q6: Analisis tambahan - Tingkat keselamatan berdasarkan Jenis Kelamin
tingkat_selamat_gender = df.groupby('Sex')['Survived'].mean() * 100

print("Tingkat Keselamatan berdasarkan Jenis Kelamin (%):")
print(tingkat_selamat_gender.round(2))
```

Ada dua hal menarik yang dapat dapat disimpulkan dari data titanic ini dimana, pertama semakin elit atau mewah kelas kapal semakin tinggi tingkat keselamatan, Kelas 1 memiliki tingkat keselamatan tertinggi (sekitar 62,96%), diikuti Kelas 2 (47,28%), dan Kelas 3 (24,24%). Ini menunjukkan bahwa penumpang di kelas utama (kaya) memiliki peluang selamat yang jauh lebih besar. Selanjutnya jenis kelamin juga terdapat gap dimana, penumpang wanita memiliki tingkat keselamatan sekitar 74,20%, sementara penumpang pria hanya 18,89%.
